```
## create white img of size 512X512
img = np.ones((512,512,3), np.int)*255
plt.imshow(img)
plt.show
```

```
img = np.ones((512,512,3), np.int8)*255
cv2.line(img, (0,500), (500,0), (150,50,80),5)
plt.imshow(img)
plt.show
```

```
img = np.ones((512,512,3), np.int16)*255
cv2.rectangle(img, (200,100), (300,300), (150,50,80),5)
plt.imshow(img)
plt.show()
```

```
img = np.ones((512,512,3), np.int16)*255
cv2.rectangle(img, (200,100), (300,300), (150,50,80),-5)
## path ="path of image"
## cv2.imwrite(path,img) to save image 
plt.imshow(img)
plt.show()
```

```
img = np.ones((512,512,3), np.int16)*255
cv2.circle(img, (250,250), (200), (150,50,80),5)
cv2.circle(img, (250,250), (10), (50,80,150),-5)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

```
img = np.ones((512,512,3), np.int16)*255
points = np.array([[10,200],[90,90],[450,200]])
cv2.polylines(img, [points],True, (25,60,50),5)
plt.imshow(img)
plt.show()
```

```
img = np.ones((512,512,3), np.int8)*255
points = np.array([[10,200],[90,90],[450,200]])
cv2.polylines(img, [points], False , (25,60,50),5)
plt.imshow(img)
plt.show()
```

```
img = np.ones((512,512,3),np.int8)*255
cv2.putText(img,"Divyam",(50,300),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,6,(25,90,78),10)
plt.imshow(img)
plt.show()
```
