```
import numpy as np
import matplotlib.pyplot as plt
```

```
### Line Plot Graph
x = [1,2,5,8,9]
y = [4,6,7,9,1]
plt.figure(figsize=(8,4))  ### width,height
plt.title("Line Plot Graph",fontsize=15)
plt.xlabel("X Axis -- >>",fontsize=12)
plt.ylabel("Y Axis -- >>",fontsize=12)
plt.grid()
plt.plot(x,y,color="purple",label="Line Graph")
plt.legend(loc=2)
plt.show()
```

```
### Line Plot Graph
x1 = [1,2,5,8,9]
y1 = [4,6,7,9,1]
y2 = [1,2,5,8,9]
x2 = [4,6,7,9,1]
plt.figure(figsize=(8,4))  ### width,height
plt.title("Line Plot Graph",fontsize=15)
plt.xlabel("X Axis -- >>",fontsize=12)
plt.ylabel("Y Axis -- >>",fontsize=12)
plt.plot(x1,y1,color="purple",label="Line Graph 1")
plt.plot(x2,y2,color="red",label="Line Graph 2")
plt.legend(loc=3)
plt.show()
```

```
### Scatter Plot
x = [1,2,5,8,9,3,4,6,7]
y = [1,2,3,4,5,6,7,8,9]
plt.figure(figsize=(8,4))  ### width,height
plt.title("Scatter Plot Graph",fontsize=15)
plt.xlabel("X Axis -- >>",fontsize=12)
plt.ylabel("Y Axis -- >>",fontsize=12)
plt.scatter(x,y,color="purple",label="Scatter Graph 1",s=100,marker="*")
plt.scatter(y,x,color="red",label="Scatter Graph 2",s=200,marker="o")
plt.legend(loc=2)
plt.show()
```

```
### Bar Plot Graph
x = [1,2,5,8,9]
y = [40,60,70,90,10]
plt.figure(figsize=(8,4))  ### width,height
plt.title("Bar Plot Graph",fontsize=15)
plt.xlabel("X Axis -- >>",fontsize=12)
plt.ylabel("Y Axis -- >>",fontsize=12)
plt.bar(x,y,color="purple",label="Bar Graph")
plt.legend(loc=2)
plt.show()
```

```
#### Histogram
np.random.seed(10)
data = np.random.randint(1,100,50)
plt.hist(data,rwidth=0.9)
plt.show()
data
```

```
#### Pie Chart
slices = [50,90,70,50,150]
lab = ["A","B","C","D","E"]
cols = ["orange","red","blue","green","purple"]
plt.pie(slices,labels=lab,colors=cols,autopct="%1.2f%%",
        explode=(0.3,0,0,0,0))
plt.show()
```

```
#### Sub Plot
plt.figure(figsize=(10,10))
t1 = np.arange(0.1,2.0,0.01)
## cos(2pi * t)
y1 = np.cos(2 * np.pi * t1)
plt.subplot(4,1,1)
plt.plot(t1,y1)

t2 = np.arange(0.1,3.0,0.01)
y2 = np.cos(np.pi * t2)
plt.subplot(4,1,2)
plt.plot(t2,y2)

t3 = np.arange(0.1,4.0,0.01)
y3 = np.sin(2 * np.pi * t3)
plt.subplot(4,1,3)
plt.plot(t3,y3)

t4 = np.arange(0.1,5.0,0.01)
y4 = np.sin(np.pi * t4)
plt.subplot(4,1,4)
plt.plot(t4,y4)
plt.show()
```
